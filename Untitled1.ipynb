```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np


from tensorflow import keras
from tensorflow.keras import layers

data, metadata = tfds.load('mnist', as_supervised=True, with_info=True)

train_data, test_data = data['train'], data['test']

def normalizer(img, label):
  img = tf.cast(img, tf.float32)
  img /= 255
  return img, label

train_data = train_data.map(normalizer)
test_data = test_data.map(normalizer)
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(32,  (3,3), input_shape=(28,28,1), activation='relu'),
    keras.layers.MaxPooling2D(2,2),

    keras.layers.Conv2D(64,  (3,3), input_shape=(28,28,1), activation='relu'),
    keras.layers.MaxPooling2D(2,2),

    keras.layers.Dropout(0.5),

    keras.layers.Flatten(),
    keras.layers.Dense(units=100, activation="relu"),
    keras.layers.Dense(units=10, activation="softmax")
])
```

```python
model.compile(
    optimizer="adam",
    loss=keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
num_train = metadata.splits['train'].num_examples
num_test = metadata.splits['test'].num_examples
```

```python
batch=32

train_data = train_data.repeat().shuffle(num_train).batch(batch)
test_data = test_data.batch(batch)
```

```python
import math
historial = model.fit(
    train_data,
    epochs=60,
    steps_per_epoch=math.ceil(num_train/batch)
)
```
